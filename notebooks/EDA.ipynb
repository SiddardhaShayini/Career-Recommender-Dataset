## Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
```

## Load the Dataset

```python
url = "https://raw.githubusercontent.com/SiddardhaShayini/Career-Recommender-Dataset/refs/heads/main/dataset/CareerRecommenderDataset.csv"
df = pd.read_csv(url)
df.head()
```

```python
# Data types and null values
df.info()
```

```python
# Summary statistics for numeric columns
df.describe()
```

## Data Cleaning

**Check for missing values**

```python
missing = df.isnull().sum()
missing[missing > 0]
```

**Convert "Yes"/"No" to binary (1/0)**

```python
df = df.applymap(lambda x: 1 if str(x).strip().lower() == 'yes' else 0 if str(x).strip().lower() == 'no' else x)
```

**Check Unique Courses and Careers**

```python
print("Unique Degree Courses:", df['Courses'].nunique())
print("Unique Career Options:", df['Career_Options'].nunique())
print("\nSample Courses:\n", df['Courses'].unique()[:10])
print("\nSample Career Options:\n", df['Career_Options'].unique()[:10])
```

## Visualizations

**Top 10 Most Common Interests**

```python
interest_columns = df.columns[:-2]  # exclude 'Courses' and 'Career_Options'

interest_sums = df[interest_columns].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=interest_sums.values, y=interest_sums.index, palette='viridis')
plt.title("Top 10 Most Common Interests")
plt.xlabel("Count")
plt.ylabel("Interest Area")
plt.show()
```

The bar chart shows the popularity of various interest areas based on their 'Count'.

**Key Observations:**

* **Researching** is the most common interest, with a count close to 1000.
* **Teaching** is the second most common interest, with a count around 750.
* Interests like **English**, **Reading**, **Solving Puzzles**, **Mathematics**, **Content Writing**, **Designing**, **Drawing**, and **Hindi** follow, with counts ranging from approximately 500 to 600.
* **Hindi** appears to be the least common among the top 10 interests listed, with a count slightly above 500.

In summary, researching and teaching are significantly more popular interests compared to the others listed in this chart.

**Heatmap of Correlation Between Interests**

```python
plt.figure(figsize=(18,14))
sns.heatmap(df[interest_columns].corr(), cmap='coolwarm', annot=False)
plt.title("Correlation Between Interests")
plt.show()
```

This heatmap visually represents the correlation between various interests, hobbies, and professions. The red diagonal indicates a perfect correlation between identical interests, while the color gradient—from blue (low correlation) to red (high correlation)—shows relationships between different interests.

Key insights:
- Interests with strong correlations can suggest patterns in preferences and behaviors.
- The clustering of colors helps identify closely related fields or hobbies.
- It can be useful for guiding career choices or understanding common interest overlaps.

```python
# Get the value counts for 'Courses' and select the top 10
top_10_courses = df['Courses'].value_counts().head(10)

plt.figure(figsize=(12, 7))
sns.barplot(x=top_10_courses.values, y=top_10_courses.index, palette='crest')
plt.title('Top 10 Most Frequent Courses')
plt.xlabel('Number of Students')
plt.ylabel('Course Name')
plt.tight_layout()
plt.show()
```

```python
# Get the value counts for 'Career_Options' and select the top 10
top_10_career_options = df['Career_Options'].value_counts().head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_career_options.values, y=top_10_career_options.index, palette='magma')
plt.title('Top 10 Most Frequent Career Options')
plt.xlabel('Number of Students')
plt.ylabel('Career Option')
plt.tight_layout()
plt.show()
```

```python
# Get top 5 courses and top 5 interests for a focused heatmap
top_5_courses_list = df['Courses'].value_counts().head(5).index.tolist()
top_5_interests_list = df[interest_columns].sum().sort_values(ascending=False).head(5).index.tolist()

# Filter data for top courses and interests
df_filtered = df[df['Courses'].isin(top_5_courses_list)]

# Calculate the mean of top interests grouped by courses
interest_course_pivot = df_filtered.pivot_table(index='Courses', values=top_5_interests_list, aggfunc='mean')

plt.figure(figsize=(12, 8))
sns.heatmap(interest_course_pivot, annot=True, cmap='YlGnBu', fmt=".2f", linewidths=.5)
plt.title('Average Presence of Top 5 Interests per Top 5 Courses')
plt.xlabel('Interest Area')
plt.ylabel('Course Name')
plt.tight_layout()
plt.show()
```

```python
# Get top 5 career options for a focused heatmap
top_5_career_options_list = df['Career_Options'].value_counts().head(5).index.tolist()

# Filter data for top career options and top interests
df_filtered_career = df[df['Career_Options'].isin(top_5_career_options_list)]

# Calculate the mean of top interests grouped by career options
interest_career_pivot = df_filtered_career.pivot_table(index='Career_Options', values=top_5_interests_list, aggfunc='mean')

plt.figure(figsize=(14, 8))
sns.heatmap(interest_career_pivot, annot=True, cmap='YlGnBu', fmt=".2f", linewidths=.5)
plt.title('Average Presence of Top 5 Interests per Top 5 Career Options')
plt.xlabel('Interest Area')
plt.ylabel('Career Option')
plt.tight_layout()
plt.show()
```

```python
# Create a cross-tabulation of Courses and Career_Options
# Filter for top 5 courses and top 5 career options for better readability
df_relation = df[df['Courses'].isin(top_5_courses_list) & df['Career_Options'].isin(top_5_career_options_list)]
course_career_crosstab = pd.crosstab(df_relation['Courses'], df_relation['Career_Options'])

if not course_career_crosstab.empty:
    plt.figure(figsize=(14, 10))
    sns.heatmap(course_career_crosstab, annot=True, cmap='Blues', fmt='d', linewidths=.5)
    plt.title('Relationship Between Top 5 Courses and Top 5 Career Options')
    plt.xlabel('Career Option')
    plt.ylabel('Course')
    plt.tight_layout()
    plt.show()
else:
    print("No data points found for the selected top 5 courses and career options combinations for cross-tabulation.")
```
