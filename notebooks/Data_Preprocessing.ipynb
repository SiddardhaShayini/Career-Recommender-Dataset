## Data Preprocessing

**Load Cleaned Dataset**

```python
import pandas as pd

# Load the cleaned dataset
url = "https://raw.githubusercontent.com/SiddardhaShayini/Career-Recommender-Dataset/refs/heads/main/dataset/cleaned_dataset.csv"
df = pd.read_csv(url)
df.head()
```

### Identify Trends, Correlations, and Patterns

**Interest Frequency Across Dataset**

```python
# Exclude 'Courses' and 'Career_Options'
interest_cols = df.columns[:-2]

# Count of people having each interest
interest_counts = df[interest_cols].sum().sort_values(ascending=False)
interest_counts.head(10)
```

**Correlation Matrix (Interests)**

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 12))
sns.heatmap(df[interest_cols].corr(), cmap="coolwarm", annot=False)
plt.title("Interest Correlation Heatmap")
plt.show()
```

This heatmap visually represents the correlation between various interests, hobbies, and professions. The red diagonal indicates a perfect correlation between identical interests, while the color gradient—from blue (low correlation) to red (high correlation)—shows relationships between different interests.

Key insights:
- Interests with strong correlations can suggest patterns in preferences and behaviors.
- The clustering of colors helps identify closely related fields or hobbies.
- It can be useful for guiding career choices or understanding common interest overlaps.



**Trends: Which interests lead to which courses?**

```python
# Group by Courses to find popular interests for each course
course_interest_summary = df.groupby('Courses')[interest_cols].sum()
course_interest_summary.T.head(10)
```
